```python
def dealing_outliers(df):
    for i in df.columns:
        if df[i].dtype == 'int64' or df[i].dtype == 'float64':
            q1 = df[i].quantile(0.25)
            q3 = df[i].quantile(0.75)
            iqr = q3 - q1
            lower_limit = q1 - 1.5 * iqr
            upper_limit = q3 + 1.5 * iqr
            df[i] = np.where(df[i] > upper_limit, upper_limit, df[i])
            df[i] = np.where(df[i] < lower_limit, lower_limit, df[i])
    return df
```

```python
def visualize_category(df, col):
    fig , ax = plt.subplots(figsize=(10,8))
    sns.countplot(x = df[col],ax=ax)
    plt.title(f'{col} Distribution',fontsize=20)
    plt.xlabel(f'{col}',fontsize=15)
    plt.ylabel('Count',fontsize=15)
    plt.xticks(rotation=75,fontsize=15)
    plt.yticks(fontsize=15)
    plt.show()
    # plt.savefig(f'{col}.png')
```

```python
def visualize_bivariate():
    fig, ax = plt.subplots(figsize=(10,8))
    sns.countplot(x = mama_df['Sentiment'], hue = mama_df['Emotion'], ax=ax)
    plt.title('Sentiment vs Emotion',fontsize=20)
    plt.xlabel('Sentiment',fontsize=15)
    plt.ylabel('Count',fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.show()
    # plt.savefig('Sentiment vs Emotion.png')
```

#### 3. Multivariate Analysis

```python
# heatmap
fig , ax = plt.subplots(figsize=(12,10))
corr = mama_df.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, mask=mask, vmax=.3, center=0,annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},ax=ax)

plt.show()
fig.savefig('Correlation chart.png')
```
